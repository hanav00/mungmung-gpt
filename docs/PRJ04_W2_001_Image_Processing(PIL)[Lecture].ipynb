````python
# 라이브러리 임포트 및 버전 확인
import os, json, random, warnings
warnings.filterwarnings('ignore')

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

import platform

def setup_korean_font():
    # 1. OS별 기본 폰트 설정
    system_os = platform.system()
    
    if system_os == "Darwin":     # Mac
        font_name = "AppleGothic"
    elif system_os == "Windows":  # Windows
        font_name = "Malgun Gothic"
    else:                         # Linux (Colab, Docker 등)
        # 나눔 폰트 설치 여부 확인 후 설정
        font_list = [f.name for f in fm.fontManager.ttflist if 'Nanum' in f.name]
        if font_list:
            font_name = font_list[0]
        else:
            # 리눅스에서 폰트가 없을 경우 설치 가이드 출력
            print("⚠️ 리눅스 환경: NanumGothic 폰트가 없습니다. 'apt-get install fonts-nanum'이 필요할 수 있습니다.")
            font_name = "sans-serif"

    # 2. Matplotlib에 적용
    plt.rcParams['font.family'] = font_name
    plt.rcParams['axes.unicode_minus'] = False
    
    print(f"✅ 현재 OS: {system_os} | 설정된 폰트: {font_name}")

# 설정 실행
setup_korean_font()
````

## 2. 이미지 처리 기초 개념

### 디지털 이미지의 구조
- 픽셀(Pixel): 이미지의 최소 단위
- 채널(Channel): 색상 정보를 담는 단위
  - RGB: Red, Green, Blue 채널 (3채널)
  - Grayscale: 흑백 이미지 (1채널)
  - RGBA: RGB + 투명도 (4채널)

### 이미지 포맷
- JPEG/JPG: 손실 압축, 작은 파일 크기
- PNG: 무손실 압축, 투명도 지원
- BMP: 무압축, 큰 파일 크기
- TIFF: 무손실, 고품질, 큰 파일 크기

## 3. PIL(Python Imaging Library) 기본 사용법

- **PIL**은 파이썬의 대표적인 이미지 처리 라이브러리이며, 현재는 **Pillow**로 발전하여 사용됨

- 이미지 **파일 포맷 변환**, **크기 조절**, **회전** 등 기본적인 이미지 처리 기능을 제공

- **ImageDraw** 모듈을 통해 도형 그리기, 텍스트 삽입 등 이미지 편집 작업이 가능

- 다양한 이미지 파일 형식(JPEG, PNG, BMP 등)을 지원하며 높은 **호환성**을 제공

- 파이썬에서 이미지 처리를 위한 필수적인 라이브러리로, 간단한 사진 편집부터 복잡한 이미지 처리까지 다양한 작업이 가능함

### 1) 기본적인 이미지 열기와 정보 확인

````python
# 이미지 열기
from PIL import Image

# 이미지 열기
img = Image.open('data/portrait.jpg')
````

````python
# 이미지 보기
plt.imshow(img);
````

````python
# 이미지 정보 출력
print(f'이미지 크기: {img.size}')  # (width, height)
print(f'이미지 형식: {img.format}')  # JPEG, PNG 등
print(f'이미지 모드: {img.mode}')   # RGB, RGBA, L(그레이스케일) 등
````

### 2) 이미지 저장하기

````python
# 다른 형식으로 저장
img.save('portrait_saved.png')
````

````python
# JPEG 품질 설정 (0-95, 기본값: 75)
img.save('portrait_saved_95.jpg', quality=95)
````

### 3) 이미지 크기 조정

````python
print(f"원본 이미지 크기: {img.size}")

# 특정 크기로 조정
# 고품질 리샘플링을 위해 Image.Resampling.LANCZOS 사용 권장 (Pillow 10+)
resized_img = img.resize((800, 600))

print(f"조정된 이미지 크기: {resized_img.size}")
````

````python
plt.imshow(resized_img);
````

````python
# 비율 유지하면서 조정
width, height = img.size
ratio = width / height
new_width = 800
new_height = int(new_width / ratio)
proportional_img = img.resize((new_width, new_height))

print(f"비율 유지 조정된 이미지 크기: {proportional_img.size}")
````

````python
plt.imshow(proportional_img);
````

### 4) 이미지 회전

````python
# 90도 회전
rotated_90 = img.rotate(90)

plt.imshow(rotated_90);
````

````python
# 임의 각도 회전 (배경색 지정)
rotated = img.rotate(45, expand=True, fillcolor='white')

plt.imshow(rotated);
````

### 5) 이미지 자르기

````python
# 특정 영역 자르기 (left, top, right, bottom)
cropped = img.crop((100, 100, 400, 400))

plt.imshow(cropped);
````

### 6) 이미지 필터 적용

````python
from PIL import ImageFilter

# 블러 필터 -> 이미지를 흐리게 만듦
blurred = img.filter(ImageFilter.BLUR)

plt.imshow(blurred);
````

````python
# 샤프닝 필터 -> 이미지를 선명하게 만듦
sharpened = img.filter(ImageFilter.SHARPEN)

plt.imshow(sharpened);
````

````python
# 엣지 검출 필터 -> 이미지의 엣지를 검출
edges = img.filter(ImageFilter.FIND_EDGES)

plt.imshow(edges);
````

### 7) 이미지 색상 조정

````python
# 흑백 변환
grayscale = img.convert('L')

plt.imshow(grayscale, cmap='gray');
````

````python
# 색상 밝기 조정
from PIL import ImageEnhance

enhancer = ImageEnhance.Brightness(img)
brightened = enhancer.enhance(1.5)  # 1.0보다 크면 밝게, 작으면 어둡게

plt.imshow(brightened);
````

### 8) 여러 이미지 합치기

````python
# 가로로 이미지 합치기
def combine_images_horizontal(images):
    total_width = sum(img.size[0] for img in images)
    max_height = max(img.size[1] for img in images)
    
    combined = Image.new('RGB', (total_width, max_height))
    x_offset = 0
    
    for img in images:
        combined.paste(img, (x_offset, 0))
        x_offset += img.size[0]
    
    return combined

combined_horizontal = combine_images_horizontal([img, rotated , edges])

plt.imshow(combined_horizontal);
````
